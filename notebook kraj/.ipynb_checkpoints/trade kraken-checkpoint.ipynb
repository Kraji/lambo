```python
import time
import cPickle

starttime=time.time()
i=0

res=k.query_public('Trades',{'pair':'XETHZEUR','since':'0'})
last= float(res['result']['last'])
data = np.array(res['result']['XETHZEUR'])[:,:3].astype(float)

time.sleep(1. - ((time.time() - starttime) % 1.))
res2=k.query_public('Trades',{'pair':'XETHZEUR'})
last_now=float(res['result']['last'])

f = open('EthEur.p','wb')
cPickle.dump(data,f)
f.close()

time.sleep(1. - ((time.time() - starttime) % 1.))
while last<last_now or i<200:
    res=k.query_public('Trades',{'pair':'XETHZEUR','since':last})
    last= float(res['result']['last'])
    f = open('EthEur.p','rb')
    A = cPickle.load(f)
    f.close()
    
    f = open('EthEur.p','wb')
    A=np.concatenate((A,np.array(res['result']['XETHZEUR'])[:,:3].astype(float)),axis=0)
    cPickle.dump(A,f)
    f.close()
    i+=1
    time.sleep(2. - ((time.time() - starttime) % 2.))

np.savetxt('LastID.txt',[last])
```
